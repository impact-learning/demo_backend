```python
d = []
for i in np.arange(7):
    t = pd.read_excel('./data.xlsx', sheetname='201{}'.format(i))
    t = t.groupby('category').sum().transpose()
    t['year'] = 2010 + i
    t['township_village'] = t.index
    o = t['township_village'].str.extract('(?P<township>.{3})(?P<village>.*)', expand=True)
    t['township'] = o.township
    t['village'] = o.village
    d.append(t)
```

```python
pd.concat(d).to_json('./data.json', orient='records', force_ascii=False)
```
