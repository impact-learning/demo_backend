```python
def avg_income(incomes):
    a = [2500, 7500, 12500, 17500, 22500, 25000]
    return np.dot(a, incomes)
```

```python
# forest_area-50mu	forest_area-50mu_100mu	forest_area-100mu_150mu	forest_area-150mu
def avg_forest_area(forest_areas):
    a = [25, 75, 125, 150]
    return np.dot(a, forest_areas)
```

```python
basic_data['avg_forest_area'] = basic_data[[
        'forest_area-50mu',
        'forest_area-50mu_100mu',
        'forest_area-100mu_150mu',
        'forest_area-150mu',
    ]].apply(avg_forest_area, axis=1)
```

```python
# planting_area-2mu	planting_area-2mu_4mu	planting_area-4mu_6mu	planting_area-6mu
def avg_planting_area(plating_areas):
    a = [1, 3, 5, 6]
    return np.dot(a, plating_areas)
```

```python
basic_data['avg_planting_area'] = basic_data[[
        'planting_area-2mu',
        'planting_area-2mu_4mu',
        'planting_area-4mu_6mu',
        'planting_area-6mu',
    ]].apply(avg_planting_area, axis=1)
```

```python
basic_data['acu_planting_area'] = 0.0
basic_data['acu_donation'] = 0.0
basic_data['acu_plants'] = 0.0
acu_planting_area = {}
acu_plantings = {}
donation = {}

for idx, row in basic_data.iterrows():
    v = ''.join([row['township'], row['village']])
    # if it is selected village
    village = project_summary[(project_summary.village == v)]
    if village.values.any():
        if len(village) >= 2:
            if row['year'] == 2012 or row['year'] == 2015:
                village = village[village.year == 2012]
            else:
                village = village[village.year == 2013]

        planting_year = village.year.values[0]
        planting_starting_mth = village.planting_starting_mth.values[0]
        planting_ending_mth = village.planting_ending_mth.values[0]
        harvest_start_year = village.year_of_maturity.values[0]
        harvest_starting_mth = village.harvest_starting_mth.values[0]
        harvest_ending_mth = village.harvest_ending_mth.values[0]
        planting_mths = np.arange(planting_starting_mth, planting_ending_mth + 1)

        planting_area = village.planting_area.values[0] / len(planting_mths)
        donated_amt = village.obtained_donation.values[0]
        num_of_plant = village.num_of_plants.values[0] / len(planting_mths)
        production = village.production.values[0]

        # donation
        if row['year'] == planting_year and row['month'] == planting_starting_mth:
            if donation.has_key(v):
                donation[v] = donation[v] + donated_amt
            else:
                donation[v] = donated_amt
            basic_data.set_value(idx, 'acu_donation', donation[v])
        elif row['year'] == planting_year and row['month'] > planting_starting_mth or\
            row['year'] > planting_year:
            basic_data.set_value(idx, 'acu_donation', donation[v])

        # num_of_plant
        if row['year'] == planting_year and row['month'] in planting_mths:
            if acu_plantings.has_key(v):
                acu_plantings[v] = acu_plantings[v] + num_of_plant
            else:
                acu_plantings[v] = num_of_plant
            basic_data.set_value(idx, 'acu_plants', acu_plantings[v])
        elif row['year'] == planting_year and row['month'] > planting_ending_mth or\
            row['year'] > planting_year:
            basic_data.set_value(idx, 'acu_plants', acu_plantings[v])

        # planting area
        if row['year'] == planting_year and row['month'] in planting_mths:
            if acu_planting_area.has_key(v):
                acu_planting_area[v] = acu_planting_area[v] + planting_area
            else:
                acu_planting_area[v] = planting_area
            basic_data.set_value(idx, 'acu_planting_area', acu_planting_area[v])
        elif row['year'] == planting_year and row['month'] > planting_ending_mth or\
            row['year'] > planting_year:
            basic_data.set_value(idx, 'acu_planting_area', acu_planting_area[v])

        if (row['year'] < planting_year) or\
            (row['year'] == planting_year and\
            row['month'] < planting_starting_mth):
            # no activity
            basic_data.set_value(idx, 'project_stage', 'noActivity')

        elif row['year'] == planting_year and\
            row['month'] in planting_mths:
            # planting
            basic_data.set_value(idx, 'project_stage', 'planting')

        elif (row['year'] == planting_year and\
            row['month'] > planting_ending_mth) or\
            (row['year'] > planting_year and row['year'] < harvest_start_year):
            # waiting for maturity
            basic_data.set_value(idx, 'project_stage', 'waitingForMaturity')

        elif row['year'] >= harvest_start_year:
            if row['month'] in np.arange(harvest_starting_mth, harvest_ending_mth + 1):
                # harvesting
                basic_data.set_value(idx, 'project_stage', 'harvesting')
            else:
                basic_data.set_value(idx, 'project_stage', 'waitingForGrowth')
        else:
            basic_data.set_value(idx, 'project_stage', 'shouldNotHappend')
    # Not selected Village
    else:
        basic_data.set_value(idx, 'project_stage', 'notSelected')
```
